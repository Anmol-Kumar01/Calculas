```python
import sympy as sp
import math
from scipy.integrate import quad
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
#Computation of a_list ,b_list and f_list 
# Here length of each list is 200

a_list = [] # a_list contain all the values a in range(2,4) with increment of 0.1
b_list = [] # b_list contain all the values b in range(1,2) with increment of 0.1
f_list =[]  # f_list contain all the values of function 'fxn' after integartion coresponding to values in a_list and b_list

for a in range(20, 40): 
  for b in range(10, 20):
    a_list.append(a/10)
    b_list.append(b/10)

a = a_list[0] # assignt the 0th index value of a_list to initiate the fxn
b = b_list[0] # assign the 0th index value of b_list initiate the fxn

def fxn(t): # functin to caluculate the perimeter of the ellipse.
  global a
  global b
  return (a**2*(math.cos(t))**2 + b**2*(math.sin(t))**2)**0.5
result, error = quad(fxn, 0, 2*math.pi )
f_list.append(result)

for i in range(1,len(a_list)): # loop for appending the values of function.
  a = a_list[i]
  b = b_list[i]
  result, error = quad(fxn, 0, 2*math.pi )
  f_list.append(result)


```

```python
# Create the 3D plot

fig = plt.figure()
ax = plt.axes(projection='3d')

# Ploting the data and the axis
ax.plot3D(a_list, b_list, f_list, 'green')

# Assigning the labels and titles
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Plot')

# Show plot
plt.show()
```

```python
# Computation of new_a_list ,new_b_list and new_f_list 
# Here length of each list is 20

new_a_list = [] # new_a_list contain all the values a in range(2,4) with increment of 0.1
new_b_list = [] # new_b_list contain all the values b in range(1,2) with increment of 0.1
new_f_list =[]  # new_f_list contain all the values of function 'fxn' after integartion coresponding to values in new_a_list and new_b_list

for a in range(20, 60):
  for b in range(1,2):
    # new_a_list.append(a/10)
    new_a_list.append(a/10-1.0)
    new_b_list.append(b/10)

a = new_a_list[0] # assign the 0th index value of new_a_list to initiate the fxn
b = new_b_list[0] # assign the 0th index value of new_b_list to initiate the fxn

def fxn(t):
  global a
  global b
  return (a**2*(math.cos(t))**2 + b**2*(math.sin(t))**2)**0.5
result, error = quad(fxn, 0, 2*math.pi )
new_f_list.append(result)

for i in range(1,len(new_a_list)): # loop for appending the values of function
  a = new_a_list[i]
  # b = new_b_list[i]
  b=1.0
  result, error = quad(fxn, 0, 2*math.pi )
  new_f_list.append(result)
print(len(new_f_list),len(new_a_list), len(new_b_list) )
print(new_a_list)
print( new_b_list)
print(new_f_list)
```

```python
# Computing the f1_new_list, f2_new_list, f3_new_list and f4_new_list which contain the values on applying formulas

b = 1 # Here we made b = 1 as constant
  
f1_new_list = [] # values yield by third formula math.pi*2*((a+b)/2)
f2_new_list = [] # values yield by second formula  math.pi*2*(((a**2+b**2)/2)**(0.5))
f3_new_list = [] # values yield by first formula math.pi*(((6*a)/5) + ((3*b)/4))
f4_new_list = [] # values yield by Ramanujan formula math.pi*((3*(a+b))- ((3*a+b)*(a+3*b))**(0.5))

# Where a is the value in the new_a_list at ith index

for i in range(len(new_a_list)): # loop for append the values after the computations of formulas into their respective lists 
  f1_new_list.append(2*math.pi*((new_a_list[i]+b)/2))
  f2_new_list.append((2*math.pi)*((new_a_list[i]**2 + b**2)/2)**0.5)
  f3_new_list.append(math.pi*(((6*new_a_list[i])/5)+((3*b)/4)))
  f4_new_list.append(math.pi*((3*(new_a_list[i]+b))- ((3*new_a_list[i]+b)*(new_a_list[i]+3*b))**(0.5)))
print(f4_new_list)
```

```python
# Ploting the comparison graph With different formula
plt.plot(new_a_list, new_f_list, label = 'Graph for Function')
plt.plot(new_a_list, f1_new_list, label = 'Graph for Formula_1')
plt.plot(new_a_list, f2_new_list, label = 'Graph for Formula_2')
plt.plot(new_a_list, f3_new_list, label = 'Graph for Formula_3')
plt.plot(new_a_list, f4_new_list, label = "Graph for Ramanujan's Formula")

plt.legend(title = 'Comparison Graph with Different Formulas:')
plt.show()
```

```python
diff_A = []
diff_B = []
diff_C = []
diff_R = []
for i in range(len(new_f_list)):
  diff_A.append((abs(new_f_list[i]-f1_new_list[i])/new_f_list[i])*100)
  diff_B.append((abs(new_f_list[i]-f2_new_list[i])/new_f_list[i])*100)
  diff_C.append((abs(new_f_list[i]-f3_new_list[i])/new_f_list[i])*100)
  diff_R.append((abs(new_f_list[i]-f4_new_list[i])/new_f_list[i])*100)
```
